```python
#zeros vs empty
#create a array with full 0 and and empty array with size

import numpy as np

a = np.zeros((3, 4))
print ( a )

b = np.empty ((2, 3))
print ( b) 

c = np.arange(0, 1000, 5) #start stop step 
print ( c )
```

```python
import numpy as np
a_mul = np.array([[1, 2, 4],
                 [1, 'hello', 3]])

print (a_mul )
print (a_mul.dtype)


d = { 1 : 0}
a_mul = np.array([[1, d, 4],
                 [1, 'hello', 3]])
print (a_mul.dtype)
# this kind of mixing data called 'object'
```

**MATHEMATICAL FUNCTIONS**


```python
import numpy as np


l1 = [1, 2, 3, 4, 5]

l2 = [[6], [7], [8], [9], [10]]
l3 = [8, 8, 8, 8, 8]
a1 = np.array(l1)
a2 = np.array(l2)
a3 = np.array(l3)
print ( a1 * a2 ) # matrix multipication

print ( a3.dot(a1) ) # dot product 
print ( np.log2(a3))


```

```python
# array manipulation

import numpy as np 

a = np.array([1, 2, 3])

a = np.append(a, [7, 8, 9])
print ( a )

a = np.insert(a, 3, [4, 5, 6])

print ( a )

b = np.array([[1, 2, 3], #(0, 1, 2)... the second parameter is col 
             [4, 5, 6]]) 
print ( np.delete(b, 1, 1) ) #detele all element in row 1st, col 1st 

print ( b.reshape(6, 1)) # reshape the array 

print ( b.flatten()) # return a copy of 1 - n array 

print ( np.mean(b) )
```
